# 3D Lung Lesion Detection using Texture and Morphological Features
## ED6001 - Medical Image Analysis Mini-Project

**Objective:** Perform segmentation + feature extraction (GLCM, LBP, wavelet) and classify benign vs malignant nodules using SVM/CNN

**Dataset:** LIDC-IDRI Dataset (Lung Image Database Consortium)

**Expected Outcome:** Report AUC, sensitivity/specificity
